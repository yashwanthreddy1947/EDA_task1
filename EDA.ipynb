**Visualizing Histogram, ScatterPlot and Heatmap by finding outliers **

```python
x=data[['Age']].head(10)
```

```python
import numpy as np
def detect_outliers(x):
  x = np.array(x,dtype=float)
  outliers = []
  threshold = 3
  mean = np.mean(x)
  std = np.std(x)
  for i in x:
    z_score = (i-mean) / std
    if np.abs(i-z_score)>threshold:
      outliers.append(i)
  return outliers
```

```python
outlier_pts = detect_outliers(x)
outlier_pts
```

```python
plt.figure(figsize=(10,6))
sns.histplot(outlier_pts)
plt.xlabel('Outliers')
plt.ylabel('Frequency')
plt.title('Histogram of detected outliers')
plt.show()
```

```python
sns.scatterplot(outlier_pts)
plt.title('Scatter plot of detected outliers')
plt.xlabel('Outliers')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.heatmap(outlier_pts)
plt.title('Heatmap of detected outliers')
plt.xlabel('Outliers')
plt.ylabel('Frequency')
plt.show()
```
